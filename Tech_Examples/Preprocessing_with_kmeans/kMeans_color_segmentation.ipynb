```python
from matplotlib.image import imread
from pathlib import Path

folder = Path("D:/Bilder/gesammelteBilder")


image = imread(folder / "solnedgang-silsanden-arrow-play.jpg")
image.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(image)
```

```python
from tkinter import N
from sklearn.cluster import KMeans, k_means
import numpy as np


X = image.reshape(-1, 3)
kmeans = KMeans(n_clusters=6).fit(X)
segmented_img = np.around(kmeans.cluster_centers_[kmeans.labels_]).astype(int) # extracts the label color - which is the centroid-coordinates in color-space
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img)
```
