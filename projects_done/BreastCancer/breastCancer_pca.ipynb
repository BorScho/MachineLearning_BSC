## Tutorial on PCA with Scikit ##

Source: https://www.datacamp.com/tutorial/principal-component-analysis-in-python

```python
from sklearn.datasets import load_breast_cancer

bc = load_breast_cancer()
X = bc.data
y = bc.target

X.shape, y.shape
```

```python
import pandas as pd
import numpy as np

y = np.reshape(bc.target, (569,1))
bc_df = pd.DataFrame(np.concatenate([X,y], axis=1))
feature_names = bc.feature_names
feature_names
```

```python
features = np.append(feature_names, 'label')
features
```

```python
bc_df.columns = features
bc_df.head()
```

```python
bc_df['label'].replace(0,'Benign', inplace=True)
bc_df['label'].replace(1, 'Malignant', inplace=True)

bc_df.tail()
```

```python
from sklearn.preprocessing import StandardScaler

x = bc_df.loc[:,feature_names].values
x = StandardScaler().fit_transform(x)
```

```python
x.shape, np.mean(x), np.std(x)
```

```python
normalized_bc = pd.DataFrame(x)
normalized_bc.tail()
```

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
principal_components_breast = pca_breast.fit_transform(x)

principal_comp_df = pd.DataFrame(principal_components_breast, columns=['comp 1', 'comp 2'])

principal_comp_df.tail()
```

```python
print(f'Explained variation per principal component: {pca_breast.explained_variance_ratio_}')
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component 1', fontsize=20)
plt.ylabel('Principal Component 2', fontsize=20)
plt.title('Principal Component Analysis of Breast Cancer Dataset', fontsize=20)

targets=['Benign', 'Malignant']
colors=['r', 'g']

for target, color in zip(targets, colors):
    keep_mask = bc_df['label'] == target
    plt.scatter(
        principal_comp_df.loc[keep_mask, 'comp 1'], 
        principal_comp_df.loc[keep_mask, 'comp 2'], 
        c = color,
        s = 50,
    )
    plt.legend(targets, prop={'size': 15})

```
